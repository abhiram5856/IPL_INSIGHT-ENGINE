```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the 18-season mega-dataset
# (Assumes ipl.csv is in your current working directory)
df_ipl = pd.read_csv('/content/ipl.csv')

# 2. Establish the Scale
print(f"✅ Data Loaded: {df_ipl.shape[0]} matches and {df_ipl.shape[1]} columns.")
print(f"📅 Seasons Analyzed: {df_ipl['season'].min()} to {df_ipl['season'].max()} (representing 2008-2025).")

# 3. Identify the "Washout Factor" (Matches with no result)
washouts = df_ipl['winner'].isnull().sum()
print(f"⛈️ Abandoned/No Result Matches: {washouts}")

# 4. Identify the "Team Drift" problem
# We combine both team columns to see all unique names across 18 years
all_teams = pd.concat([df_ipl['team1'], df_ipl['team2']]).unique()
print(f"\n🏷️ Total Unique Team Names Found: {len(all_teams)}")
print("--- List of Teams in Raw Data ---")
print(sorted(all_teams))

# 5. Quick Insight: The Toss Decision Trend
plt.figure(figsize=(10, 5))
sns.countplot(data=df_ipl, x='toss_decision', palette='coolwarm')
plt.title('IPL Toss Decisions: 2008-2025', fontsize=16, fontweight='bold')
plt.show()
```

```python
# 1. Handle Missing Values (The Washouts)
# Remove matches with no result (usually due to rain)
df_clean = df_ipl.dropna(subset=['winner']).copy()

# 2. Standardize Team Names (The "Rebrand" Mapping)
# Merges historical data for franchises that changed names
team_mappings = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru'
}

for col in ['team1', 'team2', 'toss_winner', 'winner']:
    df_clean[col] = df_clean[col].replace(team_mappings)

# 3. Comprehensive Venue Standardization (Consolidating 50+ names)
venue_standardization = {
    # Hyderabad
    'Rajiv Gandhi International Stadium, Uppal': 'Rajiv Gandhi International Stadium',
    'Rajiv Gandhi International Stadium, Uppal, Hyderabad': 'Rajiv Gandhi International Stadium',

    # Visakhapatnam (Vizag)
    'Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium, Visakhapatnam': 'ACA-VDCA Cricket Stadium',
    'Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium': 'ACA-VDCA Cricket Stadium',

    # Mumbai
    'Wankhede Stadium, Mumbai': 'Wankhede Stadium',
    'Dr DY Patil Sports Academy, Mumbai': 'Dr DY Patil Sports Academy',
    'Brabourne Stadium, Mumbai': 'Brabourne Stadium',

    # Delhi
    'Arun Jaitley Stadium, Delhi': 'Arun Jaitley Stadium',
    'Feroz Shah Kotla': 'Arun Jaitley Stadium',
    'Feroz Shah Kotla Ground': 'Arun Jaitley Stadium',

    # Bengaluru
    'M Chinnaswamy Stadium, Bengaluru': 'M.Chinnaswamy Stadium',
    'M Chinnaswamy Stadium': 'M.Chinnaswamy Stadium',
    'M. Chinnaswamy Stadium': 'M.Chinnaswamy Stadium',

    # Chennai
    'MA Chidambaram Stadium, Chepauk': 'M. A. Chidambaram Stadium',
    'MA Chidambaram Stadium, Chepauk, Chennai': 'M. A. Chidambaram Stadium',
    'MA Chidambaram Stadium': 'M. A. Chidambaram Stadium',

    # Mohali/Chandigarh
    'Punjab Cricket Association Stadium, Mohali': 'IS Bindra Stadium',
    'Punjab Cricket Association IS Bindra Stadium, Mohali': 'IS Bindra Stadium',
    'Punjab Cricket Association IS Bindra Stadium, Mohali, Chandigarh': 'IS Bindra Stadium',
    'Punjab Cricket Association IS Bindra Stadium': 'IS Bindra Stadium',

    # UAE & Others
    'Zayed Cricket Stadium, Abu Dhabi': 'Sheikh Zayed Stadium',
    'Sardar Patel Stadium, Motera': 'Narendra Modi Stadium'
}

df_clean['venue'] = df_clean['venue'].str.strip().replace(venue_standardization)

# 4. Fill Missing Cities based on Venue keywords
df_clean.loc[df_clean['venue'].str.contains('Dubai', na=False), 'city'] = 'Dubai'
df_clean.loc[df_clean['venue'].str.contains('Sharjah', na=False), 'city'] = 'Sharjah'
df_clean.loc[df_clean['venue'].str.contains('Abu Dhabi|Zayed', na=False), 'city'] = 'Abu Dhabi'
df_clean.loc[df_clean['venue'].str.contains('Chinnaswamy', na=False), 'city'] = 'Bengaluru'

# 5. Final Audit
print(f"✅ Foundation Set! Cleaned {df_clean.shape[0]} matches.")
print(f"Total Unique Teams: {df_clean['winner'].nunique()}")
print(f"Total Cleaned Venues: {df_clean['venue'].nunique()}")
print(f"Matches with missing City: {df_clean['city'].isnull().sum()}")

# Save the cleaned data for future use
df_clean.to_csv('ipl_cleaned_2008_2025.csv', index=False)
```

```python
from sklearn.preprocessing import LabelEncoder

# 1. Create a "Match Context" Feature
# Label Encoding categorical columns for the model
le_team = LabelEncoder()
le_venue = LabelEncoder()
le_toss = LabelEncoder()

# We need to fit the encoder on ALL teams so it recognizes them in both columns
all_teams = pd.concat([df_clean['team1'], df_clean['team2']]).unique()
le_team.fit(all_teams)

df_clean['team1_id'] = le_team.transform(df_clean['team1'])
df_clean['team2_id'] = le_team.transform(df_clean['team2'])
df_clean['toss_winner_id'] = le_team.transform(df_clean['toss_winner'])
df_clean['winner_id'] = le_team.transform(df_clean['winner'])

df_clean['venue_id'] = le_venue.fit_transform(df_clean['venue'])
df_clean['toss_decision_id'] = le_toss.fit_transform(df_clean['toss_decision']) # field=1, bat=0

# 2. Binary Feature: Is the Toss Winner also Team 1?
df_clean['is_team1_toss_winner'] = (df_clean['team1'] == df_clean['toss_winner']).astype(int)

# 3. Create the Target Variable
# We want to predict: "Did Team 1 win?" (1 for Yes, 0 for No)
df_clean['team1_win'] = (df_clean['team1'] == df_clean['winner']).astype(int)

# 4. Feature Selection for our Model
# We select features that are known BEFORE the result is decided
features = ['season', 'team1_id', 'team2_id', 'toss_winner_id', 'venue_id', 'toss_decision_id', 'is_team1_toss_winner']
X = df_clean[features]
y = df_clean['team1_win']

print("--- Feature Engineering Complete ---")
print(f"Features (X) shape: {X.shape}")
print(f"Target (y) shape: {y.shape}")
display(X.head())
```

```python
# 1. Sort by Season and Date for chronological accuracy
df_clean = df_clean.sort_values(by=['season', 'match_date'])

# 2. Calculate "Season Wins So Far" for Team 1
df_clean['team1_win'] = (df_clean['team1'] == df_clean['winner']).astype(int)
df_clean['team2_win'] = (df_clean['team2'] == df_clean['winner']).astype(int)

# Use shift(1) so the model only knows results BEFORE today's match
df_clean['t1_season_wins'] = df_clean.groupby(['season', 'team1'])['team1_win'].transform(lambda x: x.shift(1).cumsum().fillna(0))
df_clean['t2_season_wins'] = df_clean.groupby(['season', 'team2'])['team2_win'].transform(lambda x: x.shift(1).cumsum().fillna(0))

# 3. Add a Head-to-Head (H2H) Historical Win Rate
# This is a bit advanced, but it adds massive predictive power
def get_h2h_win_rate(row, dataframe):
    past_matches = dataframe[(dataframe['match_date'] < row['match_date']) &
                             (((dataframe['team1'] == row['team1']) & (dataframe['team2'] == row['team2'])) |
                              ((dataframe['team1'] == row['team2']) & (dataframe['team2'] == row['team1'])))]
    if len(past_matches) == 0: return 0.5 # Neutral if no history
    wins = len(past_matches[past_matches['winner'] == row['team1']])
    return wins / len(past_matches)

df_clean['h2h_rate'] = df_clean.apply(lambda row: get_h2h_win_rate(row, df_clean), axis=1)

# 4. Re-define X and y with these new "Heavyweight" features
new_features = features + ['t1_season_wins', 't2_season_wins', 'h2h_rate']
X_advanced = df_clean[new_features]
y_advanced = df_clean['team1_win']

print("--- New Advanced Features Added ---")
display(X_advanced.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Important: For time-series data, we should not "Shuffle" the split.
# We should train on early seasons and test on recent seasons.
X_train, X_test, y_train, y_test = train_test_split(X_advanced, y_advanced, test_size=0.2, shuffle=False)

# Lead Scientist Tip: Increase n_estimators (number of trees) for more stability
rf_advanced = RandomForestClassifier(n_estimators=300, max_depth=12, random_state=42)
rf_advanced.fit(X_train, y_train)

y_pred = rf_advanced.predict(X_test)
print(f"🚀 Improved Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")
```

```python
import joblib

# 1. Save the "Brains" of the project correctly
# We save the model AND the encoders so the IDs never get mixed up
model_data = {
    'model': rf_advanced,
    'features': new_features,
    'le_team': le_team,
    'le_venue': le_venue,
    'le_toss': le_toss
}
joblib.dump(model_data, 'ipl_predictor_engine.pkl')

print("✅ Model and Encoders saved! No more ID confusion.")

# 2. The Robust Predictor Function
def lead_scientist_predict(team1, team2, venue, toss_winner, toss_decision):
    """
    Handles the mapping from team names to IDs automatically
    to prevent the 'Feature Confusion' error.
    """
    # Load the engine
    data = joblib.load('ipl_predictor_engine.pkl')
    model = data['model']
    features = data['features']

    # Pre-processing the inputs into IDs
    try:
        t1_id = data['le_team'].transform([team1])[0]
        t2_id = data['le_team'].transform([team2])[0]
        v_id = data['le_venue'].transform([venue])[0]
        tw_id = data['le_team'].transform([toss_winner])[0]
        td_id = data['le_toss'].transform([toss_decision])[0]
    except ValueError as e:
        return f"❌ Error: One of the teams or venues was not in the 2008-2025 training data. {e}"

    # Calculate the 'Advanced' logic features
    is_t1_tw = 1 if team1 == toss_winner else 0

    # Look up historical H2H from our cleaned dataframe
    h2h = df_clean[(df_clean['team1'] == team1) & (df_clean['team2'] == team2)]['h2h_rate'].mean()
    if pd.isna(h2h): h2h = 0.5 # Neutral if no history

    # Construct the exact feature vector (Must match the training column order!)
    # Order: ['season', 'team1_id', 'team2_id', 'toss_winner_id', 'venue_id', 'toss_decision_id', 'is_team1_toss_winner', 't1_season_wins', 't2_season_wins', 'h2h_rate']
    input_row = [
        19,      # Current/Future Season
        t1_id, t2_id, tw_id, v_id, td_id,
        is_t1_tw,
        5, 5,    # Assuming average momentum (5 wins each) for a new match
        h2h
    ]

    # Predict
    prob = model.predict_proba([input_row])[0]

    print(f"--- 🏏 Prediction for: {team1} vs {team2} ---")
    print(f"📍 {venue}")
    print(f"🔥 {team1} Probability: {prob[1]*100:.1f}%")
    print(f"🧊 {team2} Probability: {prob[0]*100:.1f}%")

# Test with a clear match
lead_scientist_predict('Mumbai Indians', 'Chennai Super Kings', 'Wankhede Stadium', 'Mumbai Indians', 'field')
```

```python
def strategy_dashboard(team1, team2, venue, toss_winner):
    """
    Business Use Case: Pre-Match Strategy Briefing.
    Simulates the outcome of both 'Bat' and 'Field' decisions to advise the captain.
    """
    # Load model and encoders
    data = joblib.load('ipl_predictor_engine.pkl')
    model = data['model']
    le_team = data['le_team']
    le_venue = data['le_venue']
    le_toss = data['le_toss']

    # Encode constant inputs
    t1_id = le_team.transform([team1])[0]
    t2_id = le_team.transform([team2])[0]
    v_id = le_venue.transform([venue])[0]
    tw_id = le_team.transform([toss_winner])[0]
    is_t1_tw = 1 if team1 == toss_winner else 0
    h2h = df_clean[(df_clean['team1'] == team1) & (df_clean['team2'] == team2)]['h2h_rate'].mean()
    if pd.isna(h2h): h2h = 0.5

    # 1. Simulate "Field" Decision
    field_id = le_toss.transform(['field'])[0]
    input_field = [[19, t1_id, t2_id, tw_id, v_id, field_id, is_t1_tw, 5, 5, h2h]]
    prob_field = model.predict_proba(input_field)[0][1] * 100

    # 2. Simulate "Bat" Decision
    bat_id = le_toss.transform(['bat'])[0]
    input_bat = [[19, t1_id, t2_id, tw_id, v_id, bat_id, is_t1_tw, 5, 5, h2h]]
    prob_bat = model.predict_proba(input_bat)[0][1] * 100

    # --- Visualization for Stakeholders ---
    scenarios = ['Choose to Field', 'Choose to Bat']
    probabilities = [prob_field, prob_bat]

    plt.figure(figsize=(10, 5))
    colors = ['#3498db' if p == max(probabilities) else '#bdc3c7' for p in probabilities]
    sns.barplot(x=scenarios, y=probabilities, palette=colors)

    plt.axhline(50, color='black', linestyle='--', alpha=0.3)
    plt.title(f'Strategic Analysis: {toss_winner} Wins the Toss', fontsize=15, fontweight='bold')
    plt.ylabel(f'{team1} Win Probability (%)')
    plt.ylim(0, 100)

    for i, p in enumerate(probabilities):
        plt.text(i, p + 2, f'{p:.1f}%', ha='center', fontsize=12, fontweight='bold')

    plt.show()

    # Lead Scientist Recommendation
    recommendation = "FIELD" if prob_field > prob_bat else "BAT"
    print(f"💡 LEAD SCIENTIST RECOMMENDATION: If {toss_winner} wins the toss, they should choose to {recommendation}.")

# TEST THE DASHBOARD
strategy_dashboard('Chennai Super Kings', 'Mumbai Indians', 'M. A. Chidambaram Stadium', 'Chennai Super Kings')
```

```python
import shap

# 1. Initialize the SHAP Explainer with our Random Forest model
explainer = shap.TreeExplainer(rf_advanced)
shap_values = explainer.shap_values(X_test)

# Correctly slice shap_values for class 1 (positive class)
# Based on the kernel state, shap_values is a 3D array of shape (n_samples, n_features, n_classes).
# We need shap_values[:, :, 1] to get SHAP values for class 1 across all samples and features.
shap_values_class_1 = shap_values[:, :, 1]

# 2. Visualize Global Feature Importance (Why the model works generally)
print("--- 🧠 GLOBAL MODEL LOGIC (SHAP Summary) ---")
# Use the correctly sliced shap_values for class 1
shap.summary_plot(shap_values_class_1, X_test, plot_type="bar")

# 3. Visualize a SINGLE Prediction (Why did it pick Team 1 for a specific match?)
# We look at the first match in our test set
shap.initjs()
print("\n--- 🎯 LOCAL MATCH LOGIC (Match #1 in Test Set) ---")
# For force_plot, we need a 1D array of SHAP values for a single instance for a single class.
# So, shap_values[0, :, 1] selects the first sample, all features, for class 1.
shap.force_plot(explainer.expected_value[1], shap_values[0, :, 1], X_test.iloc[0,:], matplotlib=True)
```

```python
def fantasy_advisor(team1, team2, venue):
    # 1. Filter all matches involving these two teams OR this venue
    match_pool = df_clean[(df_clean['venue'] == venue) &
                          ((df_clean['team1'].isin([team1, team2])) | (df_clean['team2'].isin([team1, team2])))]

    # 2. Count Player of the Match awards in these specific conditions
    top_performers = match_pool['player_of_match'].value_counts().head(5).reset_index()
    top_performers.columns = ['Player', 'Impact_Score']

    print(f"--- 🏆 Fantasy Advisor: {team1} vs {team2} ---")
    print(f"📍 Stadium: {venue}")
    print("Suggested 'Must-Have' Players (Based on Venue Dominance):")
    display(top_performers)

    # 3. Visualization
    plt.figure(figsize=(8, 4))
    sns.barplot(data=top_performers, x='Impact_Score', y='Player', palette='YlOrBr') # Changed 'gold' to 'YlOrBr'
    plt.title('Predictive Player Impact Index', fontsize=14)
    plt.show()

# TEST: See who to pick for a match at Wankhede
fantasy_advisor('Mumbai Indians', 'Chennai Super Kings', 'Wankhede Stadium')
```
